## Complex Numbers And 2D Rotations

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_complex_number(x, c='C0'):
    lm = 1.2*np.abs(x)
    xx = np.real(x)
    yy = np.imag(x)
    plt.plot([0, xx], [0, yy], c=c)
    plt.scatter(xx, yy, c=c)
    plt.text(xx*1.04, yy*1.04, "({:.2f} + {:.2f}i)".format(xx, yy))
    plt.plot([0, 0], [-lm, lm], linewidth=1, c='k', linestyle='--')
    plt.plot([-lm, lm], [0, 0], linewidth=1, c='k', linestyle='--')
    plt.xlim([-lm, lm])
    plt.ylim([-lm, lm])
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.axis("equal")
```

### Definition of Complex Numbers

Def. i:  $i^2 = -1$,  NOTE: Does not exist in the real numbers

Def. Complex Number

### $x = a + bi,  a, b \in \mathbb{R}$

$a$ is called the "real component"

$b$ is called the "imaginary component"


Adding complex numbers is just vector addition

Ex) $(1 + 3i) + (-2 + 6i) = (-1 + 9i) $

So we might as well think of them as 2D vectors, where we plot the real component on the x-axis and the imaginary component on the y-axis.  In this case, it really is just the ordinary tip-to-tail vector addition when we think of it graphically

```python
plot_complex_number(1+3j)
```

### Multiplication of Complex Numbers

$(a + bi) (c + di) $

$= ac + bci + adi + (bi)(di)$

$= ac + (bc + ad)i + bdi^2$

$= (ac - bd) + (bc + ad)i $


### Complex Numbers As Rotation Operators

Unit norm complex number as 

$ r_{\theta} = \cos(\theta) + \sin(\theta) i$

Suppose we had a vector $\vec{v} = (x, y)$

Define $p = x + iy$


Perform the multiplication $r_{\theta} p$

$(x \cos(\theta) - y \sin(\theta)) + (x \sin(\theta) + y \cos(\theta))i $

If we interpret the complex number as a vector, the result is the same as doing a 2D counter-clockwise rotation by $\theta$!

### $ \left[ \begin{array}{cc} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{array} \right] \left[ \begin{array}{c} x \\ y \end{array} \right] $

```python
v = [1, 1]
p = v[0] + 1j*v[1]
theta = np.pi/6
rtheta = np.cos(theta) + 1j*np.sin(theta)

plot_complex_number(p)
plot_complex_number(p*rtheta)
```

Ex) If $\theta = \pi/2$ (i.e. 90 degrees)

$r_{\theta} = i$

So multiplying by $i$ is rotating 90 degrees counter-clockwise!


```python
plot_complex_number(p)
plot_complex_number(-1 + 1j)
```
